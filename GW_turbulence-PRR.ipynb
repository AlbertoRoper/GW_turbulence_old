```python
import matplotlib.pyplot as plt

plt.rc('text', usetex=False)
plt.rc('font', family='serif')
plt.rc('font', size = 24)
plt.rcParams.update({'axes.labelsize': 'x-large',
                     'axes.titlesize': 'x-large',
                     'xtick.labelsize':'x-large',
                     'ytick.labelsize':'x-large',
                     'legend.fontsize':'x-large'})

%matplotlib inline

# The functions to read the spectra, define the class run and details are in
# run.py
import run as r
import plotting as pl

import os

## Access file where data is stored
home = os.getcwd()
home = home.replace('/PRR', '')   # remove PRR in case it is the current directory already
dir0 = home + '/PRR/'
os.chdir(dir0)
```

```python
# Define every run that we want to analyze and initialize the class run
# which also reads the spectra and the time series
runs = {}                           # Initialize dictionary for the different runs to be analyzed
def initialize_run(run_name):
    run = r.run(run_name, dir0)
    runs.update({run_name:run})
    
initialize_run('K0')
initialize_run('K01_a')
initialize_run('K01_c')
initialize_run('K03')
initialize_run('K05')
initialize_run('K1')

initialize_run('M0')
initialize_run('M01_b')
initialize_run('M01_c')
initialize_run('M03')
initialize_run('M05')
initialize_run('M1')
```

```python
# We can store the runs in variables run1, run2, ... for easier access
# These are all stored in the dictionary given for runs
K0 = runs.get('K0')
K01_a = runs.get('K01_a')
K01_c = runs.get('K01_c')
K03 = runs.get('K03')
K05 = runs.get('K05')
K1 = runs.get('K1')

M0 = runs.get('M0')
M01_b = runs.get('M01_b')
M01_c = runs.get('M01_c')
M03 = runs.get('M03')
M05 = runs.get('M05')
M1 = runs.get('M1')
```

```python
# The runs that have been read can be seen as follows
print([s for s in runs.keys()])

# To check the directories that are avaible one can run
#r.read_dirs()
```

```python
# computes some characteristics of the run (as kf, OmMmax, vAmax, ...)
#ini1.characterize_run(min_col=-2, max_col=-.9)
import numpy as np
for i in runs:
    
    run = runs.get(i)
    #if run.name_run != 'ini1':
    lM = True
    if 'K' in runs.get(i).name_run:
        lM = False
    run.characterize_run(lM=lM, min_col=np.log10(5e-3), max_col=np.log10(2e-2))
    
# to see this, uncomment the following lines
print('The elements stored in ini2 are: ')
print([s for s in K0.__dict__.keys()])

# The available quantities in time_series, as a function of time are
print('\n The elements stored in time series for ini2 are: ')
print([s for s in K0.ts.keys()])
```

```python
le, cl, le_k, le_m, cl_k, cl_m = pl.legends_PRR(runs)

for i in runs:
    
    if 'K' in runs.get(i).name_run:
        runM = False; runK = True
    if 'M' in runs.get(i).name_run:
        runK = False; runM = True
    j = pl.EEM_EEK_EEGW_vs_t(runs.get(i), lhc=True, lEKmax=True, lEMmax=True,
                     EGW0=1e-10, EGW1=8e-9, t0=5e-2, t1=3e0, EK0=1e-3, EK1=2e-2, hc0=1e-7, hc1=1.5e-6,
                     ET0=1e-6, ET1=0.2, EM0=1e-3, EM1=2e-2, EMmax0=1e-2, EMmax1=2e-1, EKmax0=1e-2,
                     EKmax1=2e-1, lM=runM, lK=runK)
    
for i in range(1, j + 1):
    plt.figure(i)
    plt.legend(cl, le, loc = 'center left')
    if i == 2 or i == 5:
        plt.legend(cl_m, le_m, loc = 'lower center')
    if i == 3 or i == 6:
        plt.legend(cl_k, le_k, loc = 'lower center')
    if i == 1 or i == 4:
        plt.legend(cl, le, loc = 'center left')
```

```python
# Define every run that we want to analyze and initialize the class run
# which also reads the spectra and the time series
runs_tau = {}                           # Initialize dictionary for the different runs to be analyzed
def initialize_run(run_name):
    run = r.run(run_name, dir0)
    runs_tau.update({run_name:run})
    
initialize_run('nohel_tau01')
initialize_run('nohel_tau02')
initialize_run('nohel_tau05')
initialize_run('nohel_tau1')
initialize_run('nohel_tau2')
```
